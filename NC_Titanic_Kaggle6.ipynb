**Hanlding null values**

Calculating the median Age based on both Pclass and Sex to make the imputation more accurate. This will provide a more granular approach by considering both socio-economic class (Pclass) and gender (Sex) in estimating the median age.

```python
# Step 1: Calculate median age for each combination of 'Pclass' and 'Sex' in the training data
age_medians = dftrn.groupby(['Pclass', 'Sex'])['Age'].median()

# Step 2: Define a function to fill missing ages using the group median
def fill_age(row, medians):
    if pd.isnull(row['Age']):
        return medians[row['Pclass'], row['Sex']]
    else:
        return row['Age']

# Step 3: Apply the function to fill missing ages in both training and test data
dftrn['Age'] = dftrn.apply(lambda row: fill_age(row, age_medians), axis=1)
dftst['Age'] = dftst.apply(lambda row: fill_age(row, age_medians), axis=1)
```

```python
# Calculate mode of 'Embarked' in dftrn
embarked_mode = dftrn['Embarked'].mode()[0]

# Assign the filled values back to the 'Embarked' column directly
dftrn['Embarked'] = dftrn['Embarked'].fillna(embarked_mode)
```

```python
# Calculate median fare per 'Pclass' in dftrn
fare_medians = dftrn.groupby('Pclass')['Fare'].median()

# Fill missing fare in dftst based on the median fare of the same 'Pclass'
dftst['Fare'] = dftst.apply(
    lambda row: fare_medians[row['Pclass']] if pd.isnull(row['Fare']) else row['Fare'],
    axis=1
)
```

```python
# List of columns to convert to categorical data type
columns_to_convert = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']  # replace with your column names

# Convert each specified column to 'category' data type
for col in columns_to_convert:
    dftrn[col] = dftrn[col].astype('category')
    dftst[col] = dftst[col].astype('category')
```

```python
# Define the percentile caps for Age and Fare
age_lower, age_upper = dftrn['Age'].quantile([0.01, 0.99])
fare_lower, fare_upper = dftrn['Fare'].quantile([0.01, 0.99])

# Cap values in Age and Fare columns
dftrn['Age'] = dftrn['Age'].clip(lower=age_lower, upper=age_upper)
dftrn['Fare'] = dftrn['Fare'].clip(lower=fare_lower, upper=fare_upper)
```

By capping values at the 1st and 99th percentiles, we only modify extreme values without removing any data points. This approach is useful with limited datasets, where every row contributes to model learning.

**One Hot Encoding on all Categorical columns**

```python
# Identify the categorical columns in dftrn_prep
categorical_columns = dftrn_prep.select_dtypes(include='category').columns

# Perform one-hot encoding on the categorical columns
dftrn_prep_encoded = pd.get_dummies(dftrn_prep, columns=categorical_columns, drop_first=True)
dftrn_prep_encoded = dftrn_prep_encoded.astype(int)
```

```python
dftst.info()
```

```python
dftst_encoded = pd.get_dummies(dftst, columns=categorical_columns, drop_first=True)
dftst_encoded = dftst_encoded.astype(int)
```

```python
dftrn_prep_encoded.head()
```

```python
X_train=dftrn_prep_encoded.copy()
Y_train=dftrn['Survived']
```

```python
X_test=dftst_encoded.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Ensure X_test has the same columns as X_train by re-indexing it
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# Scale the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert the scaled arrays back to DataFrames with the original column names
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

# Perform 5-fold cross-validation
for name, model in models.items():
    scores = cross_val_score(model, X_train_res, y_train_res, cv=5, scoring='accuracy')
    print(f"{name}: Mean Accuracy = {scores.mean():.4f}, Std = {scores.std():.4f}")
```

```python
final_model=LGBMClassifier(random_state=42)
final_model.fit(X_train_res, y_train_res)
```

```python
# Make predictions on the test set
y_train_pred = final_model.predict(X_train_scaled)
y_test_pred = final_model.predict(X_test_scaled)

# Evaluate the model
# Accuracy
train_accuracy = accuracy_score(Y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.2f}")

#test_accuracy = accuracy_score(y_test, y_test_pred)
#print(f"Test Accuracy: {test_accuracy:.2f}")
```
